# Написать функцию возвращающуб столбцовую норму матрицы

решение



```python
A = np.random.randint(10, size = (4,3))
```

```python
print(A)
```

```python
la.norm(A,1)
```

```python
def norm1(A):
    #вернуть макс столбцовую норму
    max = 0
    for i in range(A.shape[1]):
      cur = np.sum(np.abs(A[:,i]))
      if (cur > max):
        max = cur
    return max
```

```python
norm1(A)
```
