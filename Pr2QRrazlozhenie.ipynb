```python
def MetodQR(A0,b0):
    n = A0.shape[0]
    A = A0.copy() 
    #b0 = b0.reshape(n,1)  
    b = b0.copy()
    E = np.eye(n)
    
    x = b*0

    Q = E.copy()
    
    for i in range(n-1):   
      a1 = A[i:,i]
      e1 = E[i:,i]          
      if (not ((a1 == a1@e1).all())):#
        X = (a1 - la.norm(a1)*e1)/la.norm((a1 - la.norm(a1)*e1))
        X = X.reshape(n-i,1)
        #print('NORM X',la.norm(X))
        U_tmp = E[i:,i:] - 2*X@X.T  
        U = E.copy()
        U[i:,i:] = U_tmp.copy()                
       # A = U@A
        A[i:,i:] = U[i:,i:]@A[i:,i:]
        #b = U@b
        b[i:,0] = U[i:,i:]@b[i:,0]
        #print('print 1',(la.norm(A[i:,i])))
        #print('print 2', A )

        ### QR method
        Q = Q@U.T

      else:
        print("ELSE")
        continue
    
    R = A.copy()
    print('QR',Q@R) #equals A0
    """
        # obratnyi hod
        for k in reversed(range(n)):
          sum = 0
          for i in range(k,n):
            sum += A[k,i]*x[i,0]
          x[k,0] = (b[k,0] - sum)/A[k,k]
        #print(x)
    """
      
    return [Q,R]
```
