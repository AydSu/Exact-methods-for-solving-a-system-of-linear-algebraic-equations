# Пункт 3, Реализуйте метод отражений: на i−ом шаге метода i−й столбец должен будет после преобразования отражения принять вид c нулями под главной диагональю:

```python
n = 3
X = np.random.rand(n)
X0 = X/la.norm(X)
X = X0.reshape(n,1)
X
```

```python
A = np.random.rand(n,n)
A
```

```python
b = np.random.rand(n).reshape(n,1)
b
#b[1:]
```

```python
def MetodOtrazheniy(A0,b0):
    n = A0.shape[0]
    A = A0.copy() 
    #b0 = b0.reshape(n,1)  
    b = b0.copy()
    E = np.eye(n)
    x = b*0
    
    for i in range(n-1):   
      a1 = A[i:,i]#perhaps A
      e1 = E[i:,i]          
      if (not ((a1 == a1@e1).all())):
        X = (a1 - la.norm(a1)*e1)/la.norm((a1 - la.norm(a1)*e1))
        X = X.reshape(n-i,1)
        U = E[i:,i:] - 2*X@X.T      
        A_tmp = U@A[i:,i:] 
        A[i:,i:] = 0
        A[i:,i:] = A_tmp
        b_tmp = U@b0[i:]
        b[i:] = b_tmp
        print('print 1',U)
        print('print 2', A_tmp )
      else:
        print("ELSE")
        continue

    # obratnyi hod
    x[n-1] = b[n-1]
    for i in reversed(range(n-1)):
      sum = 0
      for j in range(i+1,n):
        sum += A[i,j]*x[j]
      x[i] = b[i] - sum
    
    print(x)  

    return [A,b]
```

```python
def MetodOtrazheniy2(A0,b0):
    n = A0.shape[0]
    A = A0.copy() 
    #b0 = b0.reshape(n,1)  
    b = b0.copy()
    E = np.eye(n)
    
    x = b*0
    
    for i in range(n-1):   
      a1 = A[i:,i]
      e1 = E[i:,i]          
      if (not ((a1 == a1@e1).all())):#
        X = (a1 - la.norm(a1)*e1)/la.norm((a1 - la.norm(a1)*e1))
        X = X.reshape(n-i,1)
        print('NORM X',la.norm(X))
        U_tmp = E[i:,i:] - 2*X@X.T  
        U = E.copy()
        U[i:,i:] = U_tmp.copy()                
       # A = U@A
        A[i:,i:] = U[i:,i:]@A[i:,i:]
        #b = U@b
        b[i:,0] = U[i:,i:]@b[i:,0]
        print('print 1',(la.norm(A[i:,i])))
        #print('print 2', b_tmp )
      else:
        print("ELSE")
        continue

    # obratnyi hod
    for k in reversed(range(n)):
      sum = 0
      for i in range(k,n):
        sum += A[k,i]*x[i,0]
      x[k,0] = (b[k,0] - sum)/A[k,k]
    print(x)

      
    return [A,b]
```
