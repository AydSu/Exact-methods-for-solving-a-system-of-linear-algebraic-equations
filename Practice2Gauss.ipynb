```python
#@jit
def test(A0, b0):
    #global n
    A=A0.copy()
    b=b0.copy()
    n = A.shape[0]
#    print(A@x0 - b)
    for k in range(n): # k=0..n-1
        # выбор главного элемента - TODO
        b[k]/=A[k,k]
        A[k,k:]/=A[k,k]
        for i in range(k+1,n):    
            b[i] -= A[i,k]*b[k]
            A[i,:] -= A[i,k]*A[k,:]
    #    for j in range(0,n):
    #        A[i,j] -= A[i,0]*A[0,j]
    #A[1:,:] -= A[0,:]*A[1:,0:1]
#    print('after Gauss:', A .dot(x0) - b)
    x=np.zeros(n)
    x[-1] = b[-1]
    #x[n-k] = b[n-k]-   A[n-k,n-1]*x[n-1]-A[n-k,n-2]*x[n-2]-A[n-k,n-3]*x[n-3]-...A[n-k,n-k]*x[n-k]
    for k in range(n-2,-1,-1):
        x[k] = b[k]-np.sum(x[k+1:]*A[k,k+1:])
#    print('check obr. hod:', A .dot(x) - b)
#    print(x-x0)
    return x


x0 = np.ones(n) 
b= A.dot(x0)
#print(A.dot(x0)-b)
#np.max(np.abs(np.tril(test(),-1)))
xx=test(A,b)
la.norm(A.dot(xx)-b)/la.norm(b)
```
