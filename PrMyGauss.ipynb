# Метод Гаусса (обычный)

```python
def beautyGauss(A,b):
  n = A.shape[0]
  t = A[:,:] * 0
  x = b*0
  print(x)
  for k in range(n-1):#n-1
    t[k+1:,k] = A[k+1:,k]/A[k,k]
    z = t[k+1:,k].shape[0]
    
    A[k+1:,k+1:] = A[k+1:,k+1:] - t[k+1:,k]*A[k,k+1:]
    print(A[k,k+1:])   
    A[k+1:,k] = A[k+1:,k] - A[k,k]*t[k+1:,k] #A[k+1:,k] = 0  
    b[k+1:,0] = b[k+1:,0]-t[k+1:,k]*b[k,0]
    
  x[n-1,0] = b[n-1,0]/A[n-1,n-1]
  for k in reversed(range(n-1)):
    x[k,0] = (b[k,0] - np.sum(A[k,k+1:]*x[k+1:,0]) )/A[k,k]
  print('A',A)
  print('b',b)
  print('Ax',A@x)
  return x
```

# Метод Гаусса с постолбцовым выбором максимального (главного) элемента


то есть делим не на первую строку (у которой первый множитель может оказаться = 0), в оставшейся подсистеме, а на место первой - ставим строку с максимальным множителем (максимум выбирается по столбцу)

```python
def GlavGauss(A,b): 
  Eps = 0.001
  n = A.shape[0]
  t = A[:,:] * 0
  x = b*0
  print(x)
 
  for k in range(n-1):
    #max el stolbec method begin
    st_el_max = 0
    max_idx = 0
    for idx in range(k,n):
      if (A[idx,k] > st_el_max):
        st_el_max = A[idx,k]
        max_idx = idx
    if (st_el_max < Eps):
      return 'Odnoznachnogo resheniya net'
    else:
      b[k], b[max_idx] = b[max_idx], b[k]
      A[k,k:], A[max_idx,k:] = A[max_idx,k:], A[k,k:]
    #max el stolbec method end
    
    t[k+1:,k] = A[k+1:,k]/A[k,k]
    z = t[k+1:,k].shape[0]
    
    A[k+1:,k+1:] = A[k+1:,k+1:] - t[k+1:,k]*A[k,k+1:]
    print(A[k,k+1:])   
    A[k+1:,k] = A[k+1:,k] - A[k,k]*t[k+1:,k] #A[k+1:,k] = 0  
    b[k+1:,0] = b[k+1:,0]-t[k+1:,k]*b[k,0]
    
  x[n-1,0] = b[n-1,0]/A[n-1,n-1]
  for k in reversed(range(n-1)):
    x[k,0] = (b[k,0] - np.sum(A[k,k+1:]*x[k+1:,0]) )/A[k,k]
  print('A',A)
  print('b',b)
  print('A@x',A@x)
  return x
```
